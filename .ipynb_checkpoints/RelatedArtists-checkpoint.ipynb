```python
from collections import defaultdict, deque
import pandas as pd
import time 
from urllib.parse import quote
import requests 

'''Functions for querying Spotify API'''
def build_query(artist):     
    return quote(artist)

def get_artist_id(artist):
    '''Takes the encoded artist name from build_query() to fetch artist id from API and return id'''
    
    r = requests.get("https://api.spotify.com/v1/search?q={0}&type=artist".format(artist), 
                     headers={'Authorization': 'Bearer {}'.format(token['access_token'])})
    if r.status_code == 200:
        return r.json()['artists']['items'][0]['id']
    else:
        raise IOError('Spotify API request unsuccessful. Status code:{}'.format(r.status_code))
        
def get_related_artists(artist_id):
    '''Takes the artist_id returned from get_artist_id() and returns a dictionary
       where the keys are the related artist ids and the values are the artist names'''
    
    r = requests.get("https://api.spotify.com/v1/artists/{0}/related-artists".format(artist_id),
                    headers={'Authorization': 'Bearer {}'.format(token['access_token'])})
    if r.status_code == 200:
        related_artists = {}
        for artist_dict in r.json()['artists']:
            related_artists[artist_dict['id']] = artist_dict['name']
        return related_artists
    else:
        raise IOError('Spotify API request unsuccessful. Status code:{}'.format(r.status_code))
        
def get_top_tracks(artist_id):
    '''Takes the artist_id returned from get_artist_id() and returns a dictionary 
       where keys are track ids and values are the track name'''
    
    r = requests.get("https://api.spotify.com/v1/artists/{0}/top-tracks?country=US".format(artist_id),
                    headers={'Authorization': 'Bearer {}'.format(token['access_token'])})
    if r.status_code == 200:
        top_tracks = {}
        for tracks_dict in r.json()['tracks']:
            top_tracks[tracks_dict['id']] = tracks_dict['name']
#         print ("get_top_tracks:", top_tracks)
        return top_tracks
    else:
        raise IOError('Spotify API request unsuccessful. Status code:{}'.format(r.status_code))
        
    
def get_track_info(track_id):
    r = requests.get("https://api.spotify.com/v1/audio-analysis/{0}".format(track_id), 
                     headers={'Authorization': 'Bearer {}'.format(token['access_token'])})
    if r.status_code == 200:
        return r.json()
    else:
        raise IOError('Spotify API request unsuccessful. Status code:{}'.format(r.status_code))

def get_related_artist_tracks(artist_id, return_ids = False):
    artist_tracks_names = defaultdict(list)
    artist_tracks_ids = defaultdict(list)
    
    related_artists = get_related_artists(artist_id)
    for related_artist_id, related_artist_name in related_artists.items():
        top_tracks = get_top_tracks(related_artist_id)
        if not top_tracks:
            pass
        else:
            artist_tracks_ids[related_artist_id] = top_tracks.keys()
            artist_tracks_names[related_artist_name] = top_tracks.values()
    if return_ids:
        return artist_tracks_ids
    return artist_tracks_names

def generate_artists(artist_name, num_artists):
    query = build_query(artist_name)
    artist_id = get_artist_id(query)
    Q = deque()
    visited = []
    Q.append((artist_id, artist_name))
    while len(Q) != 0 and len(visited)<num_artists:
        v = Q.popleft()
        print (v)
        if v not in visited:
            visited.append(v)
            related_artists = get_related_artists(v[0])
            Q.extend(related_artists.items())
        time.sleep(3)
    return visited

# changed function to fetch tracks after
# artist list was populated
def generate_songs(artist_list):
    songs = []   
    refresh_token = 0
    for art_id, art_name in artist_list:
        if refresh_token%20 == 0:
            token = authorization.get_token()
        top_tracks = get_top_tracks(art_id)
        if not top_tracks:
            continue
        else:
            for song_id, song_name in top_tracks.items():
                songs.append({'artist':art_name, 'song':song_name, 
                              'artist_id':art_id, 'song_id':song_id})
        time.sleep(3)
    return pd.DataFrame(songs)
```

```python
def build_dataset(list_of_bands, num_artists):
    '''Takes a list of band names to find related artists. 
       Will search for related artists until num_artists is reached for each artists.
       Returns pandas df of song, song_id, artist, and artist_id.'''
    first_artist = list_of_bands.pop()
    df = generate_df(first_artist, num_artists)
    for band in list_of_bands:
        df_temp = generate_df(band, num_artists)
        df = df.append(df_temp)
        time.sleep(3)
        token = authorization.get_token()
    df.drop_duplicates(inplace=True)
    df.index = pd.RangeIndex(len(df.index))

    return df
```

```python
def extract_features(sample):
    max_loudness = 0
    for seg in sample['segments']:
        loud = seg['loudness_max']
        if loud < max_loudness:
            max_loudness = loud

    duration = sample['track']['duration']
    loudness = sample['track']['loudness']
    dyn_range = loudness - max_loudness
    tempo = sample['track']['tempo']
    key = sample['track']['key'] 
    time_sig = sample['track']['time_signature']
    fade_in_length = sample['track']['end_of_fade_in']
    fade_out_length = duration - sample['track']['start_of_fade_out']
    total_fade = fade_in_length + fade_out_length
    return {'duration': duration, 'loudness': loudness, 
            'dyn_range': dyn_range, 'tempo': tempo, 
            'key': key, 'time_sig': time_sig, 
            'total_fade': total_fade, 
            'song_id':sample['track_info']['song_id']}
```

```python
rows = []
for doc in db.coll.find():
    rows.append(extract_features(doc))
```
